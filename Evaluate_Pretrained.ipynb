```python
def df_to_tabularx_latex(df, caption="", label="tab:", as_percentage=False, metric_only=False):
    import pandas as pd

    def escape_latex(s):
        """Escape LaTeX special characters in strings: %, _, and others if needed."""
        return str(s).replace("_","\\_").replace('%', '\\%')

    def safe_format_percentage(x):
        try:
            val = float(x)
            return f"{val * 100:.2f}\\%"
        except:
            return escape_latex(x)

    def safe_format_float(x):
        try:
            val = float(x)
            return f"{val:.2f}"
        except:
            return escape_latex(x)

    formatted_df = df.copy()

    if metric_only and 'metric' in formatted_df.columns:
        if as_percentage:
            formatted_df['metric'] = formatted_df['metric'].apply(safe_format_percentage)
        else:
            formatted_df['metric'] = formatted_df['metric'].apply(safe_format_float)
    else:
        for col in formatted_df.columns:
            if pd.api.types.is_numeric_dtype(formatted_df[col]):
                if as_percentage:
                    formatted_df[col] = formatted_df[col].apply(safe_format_percentage)
                else:
                    formatted_df[col] = formatted_df[col].apply(safe_format_float)
            else:
                formatted_df[col] = formatted_df[col].apply(escape_latex)

    # Bold the model column if it exists (also after escaping)
    if 'model' in formatted_df.columns:
        formatted_df['model'] = formatted_df['model'].apply(lambda x: f"\\textbf{{{escape_latex(x)}}}")

    # Bold and escape headers
    columns = formatted_df.columns.tolist()
    header_row = " & ".join(f"\\textbf{{{escape_latex(col)}}}" for col in columns) + " \\\\"

    latex_str = "\\begin{table}[h!]\n\\centering\n"
    latex_str += f"\\caption{{{caption}}}\n"
    latex_str += f"\\label{{{label}}}\n"
    latex_str += "\\begin{tabularx}{\\textwidth}{%s}\n" % ("l" * len(columns))
    latex_str += "\\toprule\n"
    latex_str += header_row + "\n"
    latex_str += "\\midrule\n"

    for _, row in formatted_df.iterrows():
        row_values = " & ".join(str(v) for v in row.values) + " \\\\"
        latex_str += row_values + "\n"

    latex_str += "\\bottomrule\n"
    latex_str += "\\end{tabularx}\n"
    latex_str += "\\end{table}"

    return latex_str
```

```python
import torch
import torch.nn as nn
from skorch import NeuralNetRegressor, NeuralNetClassifier
from skorch.callbacks import EpochScoring, Callback
from sklearn.metrics import r2_score, accuracy_score, confusion_matrix
from skorch.dataset import ValidSplit
from collections import defaultdict
import argparse
import pandas as pd
import numpy as np
from easydict import EasyDict as edict

#parser = argparse.ArgumentParser(description="Train regression or classification model with skorch")
#parser.add_argument('--arch', type=str, required=True, help='Architecture to extract features from')
#parser.add_argument('--method', type=str, choices=['regression', 'classification'], required=True, help='Task type')
#args = parser.parse_args()
args = edict()
args.arch = "vit_b_16"
args.method = "classification"
# Load data


# parser = argparse.ArgumentParser(description="Train regression or classification model with skorch")
# parser.add_argument('--arch', type=str, required=True, help='Architecture to extract features from')
# parser.add_argument('--method', type=str, choices=['regression', 'classification'], required=True, help='Task type')
# args = parser.parse_args()

# Load data
data = torch.load("idsprites/idsprites.pth")
model = args.arch
reps = torch.load(f"idsprites/idsprites_images_feats_{model}.pth")

torch.manual_seed(42)
X = reps
print("Recentering representations!")
X = X - X.mean(dim=0)
print("Normalizing reps!")
X = torch.nn.functional.normalize(X, p=2.0, dim=1, eps=1e-12)

# Model definitions
class LinearRegressionModel(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.linear = nn.Linear(input_dim, 1)

    def forward(self, x):
        return self.linear(x)

class LinearClassificationModel(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.linear = nn.Linear(input_dim, num_classes)

    def forward(self, x):
        return self.linear(x)

# Custom callback to store confusion matrices for the best model
class BestConfusionMatrixCallback(Callback):
    def __init__(self):
        self.best_cm = None
        self.best_score = -np.inf

    def on_epoch_end(self, net, dataset_train=None, dataset_valid=None, **kwargs):
        valid_loader = net.get_iterator(dataset_valid, training=False)
        y_true, y_pred = [], []
        for batch in valid_loader:
            Xi, yi = batch
            preds = net.predict(Xi)
            y_true.extend(yi.cpu().numpy())
            y_pred.extend(preds)
        y_true = np.array(y_true)
        y_pred = np.array(y_pred)
        current_acc = accuracy_score(y_true, y_pred)
        if current_acc > self.best_score:
            self.best_score = current_acc
            self.best_cm = confusion_matrix(y_true, y_pred)

input_dims = {    
    "vit_b_16": 768,    
    "vit_b_32": 768,
    "vit_l_16": 1024,
    "vit_l_32": 1024
}

all_results = dict()
attributes = ['shape','scale','orientation','x','y']

for i, attribute in enumerate(attributes):
    y = data['latent_ids'][:, i]

    cm_callback = BestConfusionMatrixCallback() if args.method == 'classification' else None

    if args.method == 'classification':
        y = y.long()  # Targets for classification must be integer
        model_class = LinearClassificationModel
        loss_fn = nn.CrossEntropyLoss
        accuracy_cb = EpochScoring(
            accuracy_score, name="accuracy", lower_is_better=False, on_train=False
        )
        net = NeuralNetClassifier(
            model_class,
            module__input_dim=input_dims[model],
            module__num_classes=len(y.unique()),
            max_epochs=1,
            lr=0.01,
            device="cuda",
            optimizer=torch.optim.Adam,
            criterion=loss_fn,
            batch_size=2048,
            train_split=ValidSplit(0.2),
            callbacks=[accuracy_cb, cm_callback],
            iterator_train__shuffle=True,
            verbose=1
        )
    else:
        y = y.float()
        model_class = LinearRegressionModel
        loss_fn = nn.MSELoss
        r2_cb = EpochScoring(
            r2_score, name="r2", lower_is_better=False, on_train=False
        )
        net = NeuralNetRegressor(
            model_class,
            module__input_dim=input_dims[model],
            max_epochs=1,
            lr=0.01,
            device="cuda",
            optimizer=torch.optim.Adam,
            criterion=loss_fn,
            batch_size=2048,
            train_split=ValidSplit(0.2),
            callbacks=[r2_cb],
            iterator_train__shuffle=True,
            verbose=1
        )

    print(f"Fitting the {args.method} for '{attribute}'")
    net.fit(X, y if args.method == 'regression' else y)

    results = defaultdict(list)
    metric_key = 'r2' if args.method == 'regression' else 'accuracy'
    for metrics in net.history:
        for key in ['epoch','train_loss','valid_loss', metric_key]:
            results[key].append(metrics[key])

    df = pd.DataFrame(results)

    if args.method == 'classification' and cm_callback.best_cm is not None:
        best_cm_df = pd.DataFrame(cm_callback.best_cm)
        best_cm_df.to_csv(f"confusion_matrix_best_{model}_{attribute}.csv", index=False)

    all_results[attribute] = df

# Save results
for attr, df in all_results.items():
    df.to_csv(f"results_{args.method}_{model}_{attr}.csv", index=False)

print("Training complete.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def plot_confusion_matrix(csv_path, title=None, save_path=None):
    cm = pd.read_csv(csv_path).values
    cm_percent = cm / cm.sum(axis=1, keepdims=True) * 100

    plt.figure(figsize=(20, 16))
    ax = sns.heatmap(
        cm_percent,
        annot=True,  # Show annotations again
        fmt=".1f",
        annot_kws={"size": 6},  # Smaller annotation text size
        cmap="Blues",
        cbar_kws={'label': 'Percentage'},
        linewidths=0.1,
        square=False
    )
    ax.set_xlabel('Predicted')
    ax.set_ylabel('True')
    if title:
        ax.set_title(title, fontsize=18)
    ax.tick_params(axis='x', rotation=90)
    ax.tick_params(axis='y', rotation=0)
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300)
    plt.show()

# Example usage:
# plot_confusion_matrix("confusion_matrix_best_vit_b_16_shape.csv", title="Confusion Matrix for shape", save_path="shape_cm.png")

# Example usage:
for model in ['vit_b_16','vit_b_32','vit_l_16','vit_l_32']:
    for attribute in ['shape','scale','orientation','x','y']:
        plot_confusion_matrix(f"confusion_matrix_best_{model}_{attribute}.csv", title=f"Confusion Matrix for {attribute} - {model}", save_path=f"{model}_{attribute}_cm.png")
```
