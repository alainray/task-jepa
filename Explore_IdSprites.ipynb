```python
import torch
from idsprites.visualization import draw_batch

ds = torch.load("idsprites/idsprites.pth", map_location="cpu")
```

## Exploring all shapes

```python
indices = [i*(14**4) for i in range(54)]
formas = ds['images'][indices].repeat(1,3,1,1).float()
draw_batch(formas, num_images=len(formas))
```

## Scale

```python
sample_shape = 53
indices = [sample_shape*(14**4) + i*(14**3) for i in range(14)]
formas = ds['images'][indices].repeat(1,3,1,1).float()
draw_batch(formas, num_images=len(formas))
```

## Orientation

```python
sample_shape = 4
indices = [sample_shape*(14**4) + i*(14**2) for i in range(14)]
formas = ds['images'][indices].repeat(1,3,1,1).float()
draw_batch(formas, num_images=len(formas))
```

## X - Coordinate

```python
sample_shape = 4
indices = [sample_shape*(14**4) + i*(14) for i in range(14)]
formas = ds['images'][indices].repeat(1,3,1,1).float()
draw_batch(formas, num_images=len(formas))
```

## Y Coordinate

```python
sample_shape = 4
indices = [sample_shape*(14**4) + i for i in range(14)]
formas = ds['images'][indices].repeat(1,3,1,1).float()
draw_batch(formas, num_images=len(formas))
```
